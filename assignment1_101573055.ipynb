```python
#Dictionary "Name" workout minutes(yoga, running, weightlifting)
workout_stats = { 
    "Alex": (30,45,20),
    "Jamie": (35,50,25),
    "Taylor": (40,55,30)
}
print(workout_stats)
```

```python
friends = list(workout_stats.keys())
for friend in friends:
    yoga = workout_stats[friend][0]
    running = workout_stats[friend][1]
    lifting = workout_stats[friend][2]

    total = yoga + running + lifting
    workout_stats[friend + "_Total"] = total

print("Updated dictionary with totals")
print(workout_stats)
```

```python
# 2D nested list storing workout minute

workout_list = [
    list(workout_stats["Alex"]),
    list(workout_stats["Jamie"]),
    list(workout_stats["Taylor"])
]

print(workout_list)
```

```python
# Slice the workout_list

print("Yoga and running minute")
for row in workout_list:
    print(row[0:2])

# Weightlifting
print("Weightlifting minutes (last two friends):")
for row in workout_list[-2:]:
    print(row[2])
```
